## **Distribuciones multivariadas discretas**

$$f(x_1,x_2,...x_n)=P[X_1=x_1;X_2=x_2;...;X_n=x_n]$$

## **Distribucion de probabilidad marginal**

$$f_{X_i }(x_i )=P[X=x_i ]=\sum_{R_{X_i }}f(x_1,x_2,…,x_n )$$

## **Esperanza y covarianza**

$$E[X]=
\begin{equation}
\begin{bmatrix}
E[X_1]\\
E[X_2]\\
\vdots\\
E[X_n]
\end{bmatrix}
\end{equation}$$

$$Cov(X ,Y)= E [( X - E[X ])(Y - E[Y ])]$$

$$Cov(X ,Y)= E[XY]-E[X]E[Y]$$




**Ejemplo:** La probabilidad de que la bombilla de un proyector durará menos de 40 horas de uso continuo es de 0.30, entre 40 a 80 horas de uso continuo, su probabilidad es de 0.50 y para  más de 80 horas de uso continuo, su probabilidad es de 0.20 .

$X_1$:menos de 40 horas      ,      $p_1=0.30$

$X_2$:entre 40 a 80 horas    ,      $p_2=0.50$

$X_3$:mas de 80 horas               $p_3= 0.20$

Para   $x_1+x_2+x_3=n$   ,    $p_1+p_2+p_3=1$

 $f(x_1,x_2,x_3 )=P[X_1=x_1  , X_2=x_2  , X_3=x_3  ]=\frac{n!}{x_1 !x_2 !x_3 !} p_1^{x_1}p_2^{x_2} p_3^{x_3}$

 siendo  $n=8$ ; $x_1=2,x_2=5,x_3=1$

$f(x_1,x_2,x_3 )=P[X_1=2  , X_2=5  , X_3=1  ]=\frac{8!}{2 !5 !1 !} 0.3^{x_1}0.5^{x_2} 0.2^{x_3}$

## **Distribución normal bidimensional**

Si el vector aleatorio

$X =
\left[\begin{array}{lcr}
 X_1\\
 X_2
\end{array}\right]
$

sigue una distribución normal bivariante con vector de medias

$$ $$

$\mu =
\left[\begin{array}{lcr}
 \mu_1\\
 \mu_2
\end{array}\right]
$

$$ $$

Tiene función de densidad

$$f(X)=\frac{1}{2\pi|\Sigma|^{1/2}}exp \left ( -\frac{1}{2}(X-\mu)\Sigma^{-1}(X-\mu)'   \right  )$$

$$ $$

donde:

$$\Sigma=
\left[\begin{array}{lcr}
Var(X_1) & Cov(X_1,X_2)\\
Cov(X_1,X_2) & Var(X_2)
\end{array}\right]
$$

$$ $$


$$\Sigma=
\left[\begin{array}{lcr}
\sigma^2_1 & \rho \sigma_1\sigma_2\\
\rho \sigma_1\sigma_2 & \sigma^2_2
\end{array}\right]
$$

$$ $$

$D =
\left[\begin{array}{lcr}
\lambda_1 & 0\\
 0 & \lambda_2
\end{array}\right]
$

$$ $$

$\Sigma^{1/2}=PD^{1/2}P^t$



$|\Sigma|=\sigma^2_1\sigma^2_2(1-\rho^2)$



**Ejemplo**


$\mu_x=0\\ \sigma_x=1$


$\mu_y=0\\ \sigma_y=1$

$$ $$

$f(x,y)=\frac{1}{2\pi}e^{-\frac{1}{2}(x^2+y^2)}$

```r
mu = c(0, 0)
mu
Sigma = matrix(c(1, 0.9, 0.9, 1), nrow = 2, ncol = 2)
Sigma
```

```r
rmvn.eigen =function(N, mu, Sigma) {
                n = length(mu)
                av = eigen(Sigma, symmetric = TRUE)
                L = av$values
                P = av$vectors
                D = diag(sqrt(L))
                Q = P %*% D %*% t(P)
                Z = matrix(rnorm(N*n), nrow = N, ncol =n)
                mx=matrix(mu, nrow=N, ncol=n, byrow = TRUE)
                X = Z %*% Q + mx
                X
                }
```

```r
N=1000
set.seed(147)
X =rmvn.eigen(N, mu, Sigma)
plot(X, xlab = "x", ylab = "y", pch = 20)
```
