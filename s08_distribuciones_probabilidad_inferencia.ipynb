# **La Distribución Gamma**

image.png

Para graficar la función Gamma se usa el siguiente formato

$$dgamma(x, shape, scale )$$


$$pgamma(q,shape, scale)$$


$rate =\beta$

$shape=\alpha$

$scale=1/rate$

$$\lambda=1/\beta$$


es decir


$$dgamma(x, alpha, lambda )$$


$$pgamma(q,alpha, lambda)$$


## **Grafica de la función Gamma**

**Ejemplo:** Graficar la función Gamma con $\alpha=4$ y $\beta=5$

```
x=seq(0,70,0.01)
y=dgamma(x,4,1/5)
plot(x,y,type='l',col='blue',lwd=2)
grid()
```

```
x=seq(0,20,0.01)
y=dgamma(x,3,1/2)
plot(x,y,type='l',col='blue',lwd=2)
```

**Ejemplo:** Graficar la función Gamma con $\alpha=1;\alpha=2;\alpha=4$ y $\beta=1$

```
x=seq(0,12,0.01)
y=dgamma(x,1,1)
y1=dgamma(x,2,1)
y2=dgamma(x,4,1)
#y3=dgamma(x,9,3)
plot(x,y,type='l',col='blue',lwd=2)
lines(x,y1,type='l',col='red',lwd=2)
lines(x,y2,type='l',col='green',lwd=2)
#lines(x,y3,type='l',col='black',lwd=2)
grid()
legend("topright", c("alpha =1; beta = 1", "alpha =2; beta = 1", "alpha =4; beta = 1"),
       lty = 1, col = c("blue", "red", "green"))
```

## **Cálculo de probabilidades**

**Ejemplo:**  Supóngase que el intervalo de tiempo X necesario para efectuar una comprobación periódica de mantenimiento, de acuerdo a la experiencia, en un dictáfono, sigue una Distribución Gamma con α=3 y β=2 (minutos). Calcule la probabilidad de que un mecánico necesite entre 10 y 20 minutos, para revisar un dictáfono.

**Solución:**

$P[10 \leq X \leq 20]=P[X\leq 20]-P[X\leq 10]$

```
pgamma(20,3,1/2)-pgamma(10,3,1/2)
```

##**Grafica de la distribución t-Student**

Graficar la función t-Student para $r=2;r=5; r=23$

```
x=seq(-5,5,0.01)
y=dnorm(x,0,1)
y1=dt(x,2)
y2=dt(x,5)
y3=dt(x,23)
plot(x,y,type='l',col='black',lwd=3)
lines(x,y1,type='l',col='blue',lwd=2)
lines(x,y2,type='l',col='green',lwd=2)
lines(x,y3,type='l',col='red',lwd=2)
legend("topright", c("N(0,1)", "t(2)", "t(5)","t(23)"),
       lty = 1, col = c("black","blue", "green", "red"))
```
