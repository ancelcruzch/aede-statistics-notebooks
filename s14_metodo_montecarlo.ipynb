# **El Método de Montecarlo**

Habitualmente se utiliza la denominación Monte Carlo para describir cualquier método de cálculo que utilice números aleatorios pero autores como Rubinstein (1981), Ripley (1987) y Lewis y Orav (1989), subrayan la conveniencia de utilizar este término sólo para usos específicos como la "integración Monte Carlo" o el "muestreo Monte Carlo", y utilizar el término "simulación estocástica" para los experimentos que hacen uso de series de números aleatorios generadas por ordenador.


## **Integración numérica usando el Método de Montecarlo**


Cálculo de integrales por Montecarlo En el caso que nos ocupa, se desea estimar la integral de una función $g$ continua. Esta integral puede verse como el cálculo del valor esperado de la función g cuando se aplica a una variable aleatoria con distribución uniforme.

Supongamos que el intervalo de integración es $[0, 1]$ y sea $X_1,X_2,.…, X_M$  una muestra de variables aleatorias,
independientes con distribución uniforme en el intervalo $[0, 1]$, entonces

$$\displaystyle\int_{0}^{1}g(x)dx=E(g(x))$$

$$\displaystyle\int_{0}^{1}g(x)dx=\frac{1}{M}\sum_{i=1}^Mg(x_i)$$

Cualquier  integral sobre el intervalo $[a, b]$ se puede transformar a una integral sobre el intervalo [0, 1]  con el siguiente cambio de variable  $x = a + (b - a)u$, entonces


$$\int_{a}^{b}g(x)dx=\frac{b-a}{M}\sum_{i=1}^Mg(a+(b-a)u_i)$$

**Ejemplo**

Hallar

$$\int_0^1x^2dx$$

```
M=3000
set.seed(123)
x=runif(M,0,1)
x=round(x,4)
```

```
gx=x^2
gx=round(gx,4)
```

```
(1/M)*sum(gx)
```

# **Montecarlo para inferencia**

Los métodos de Monte Carlo se pueden aplicar para estimar parámetros de la distribución de muestreo de una estadística, error cuadrático medio (MSE), percentiles u otras cantidades de interés. Los estudios de Monte Carlo pueden diseñarse para evaluar la probabilidad de cobertura para intervalos de confianza, para encontrar una tasa de error empírica Tipo I de un procedimiento de prueba, para estimar el poder de una prueba y para comparar el desempeño de diferentes procedimientos para un problema dado.

Supongamos que $X_1,. . ., X_n$ es una muestra aleatoria de la distribución de $X$. Un
estimador $\hat{\theta}$ para un parámetro $\theta$ es una función $n$ variables
$\hat{\theta} = \hat{\theta} (X_1,..., X_n)$

**Ejemplo**

Si  $X_1, X_2$ son dos variables aleatorias identicamente distribuidas como una distribucion normal estandar, estimar la diferencia de medias  $E[X_1-X_2]$

**Solución:**


$g(X_1,X_2)=|X_1-X_2|$

Generamos $m$ muestras aleatorias $x^j=(x_1^j,x_2^j)$ de variables aleatorias normales asi

$\hat{\theta}^j=g_j(x_1,x_2)=|x_1^j-x_2^j|$

$\hat{\theta}=\frac{1}{m}\sum\limits_{j=1}^m\hat{\theta}^j$

```
m = 5
g = numeric(m)
for (i in 1:m) {
set.seed(i)
x = round(rnorm(2),2)
print(x)
g[i] = abs(x[1] - x[2])
print(g[i])
}
est = mean(g)
est
```

## **Estimacion del error estandar de la media**


El error estandar de la media muestral $\overline X$ de tamaño $n$ esta dado por $\frac{\sigma}{\sqrt{n}}$, donde

\begin{equation}
\hat {\sigma^2}=\frac{1}{n}\sum\limits_{i=1}^n(x_i-\overline x)^2
\end{equation}

```
ErrorStandar=sqrt(sum((g - mean(g))^2)) / m
ErrorStandar
```
