## **El método de la transformada invers**a



Este método generalmente se utiliza para distribuciones de probabilidad que tienen una distribución acumulada sencilla de integrar y que se pueda obtener de forma cerrada, algunas variables aleatorias a las que se puede aplicar este método son: la uniforme, la exponencial, la de Weibull, la Triangular, la de Bernoulli y otras.

#### **Teorema**

Sea $X$ una variable aleatoria continua con función de distribución de probabilidad acumulada $F$. Si $U \thicksim U(0; 1)$, entonces la valeatoria continua $x = F^{−1} (U)$  tiene como función de distribución $F$.


Su algoritmo está basado en que la distribución acumulada que es la probabilidad de que $X$ tome un valor menor o igual que $x$ y los números aleatorios $R$ están distribuidos entre 0 y 1. La distribución acumulada o función de distribución, $F(x)$, de una variable aleatoria $X$ es definida para cada número real de la siguiente forma: $F(x) = P(X\leq x)$ para $-∞ < x < ∞ $, donde $P(X \leq x)$ es la probabilidad asociada con el suceso $X \leq x$.

**Algoritmo del método de la transformada inversa**

**Paso 1:** Dada una función de probabilidad $f(x)$ para una variable aleatoria $X$ obtener la distribución de probabilidad acumulada integrando $f(t)$ en el intervalo de $<-\infty, x]$

**Paso 2:** Generar un número aleatorio $R \thicksim U(0,1)$

**Paso 3:** Como $F(x)$ y R toman valores entre 0 y 1, se sustituye la función acumulada por el número aleatorio, en seguida se despeja x quedando así en función de R.

**Ejemplo:**


Dada la siguiente distribución especificada por la función



$$
f(x)=\left\{
\begin{array}{ll}
\frac{1}{b-a} , \ si \ a \leq x \leq b\\
0 \ , \  en \ otro \ caso
\end{array}
\right.
$$


Aplicando la transformada inversa

1. Hallamos $F(x)$


$$
F(x)=P[X \leq x]=\int_{-\infty}^x f(t)dt=\left\{
\begin{array}{lll}
0, \ si \ x < a\\
\frac{x-a}{b-a} , \ si \ a \leq x < b\\
1 \ , \  si \ x \geq b
\end{array}
\right.
$$


$$
F(x)=\int_{a}^x f(t)dt=\frac{x-a}{b-a}
$$

2. Generamos el número aleatorio  $R$


$$R \thicksim U(0,1)$$


3. $R=F(x)$
    
$$
R=\int_{a}^x f(t)dt=\frac{x-a}{b-a}\\
x=a+R(b-a)
$$    
    
Se aplico la transforma inversa a la distribución uniforme $U(a,b)$.



**Ejemplo:**

Para obtener un valor aleatorio  $U(3,6)$  generamos un número aleatorio $R$ uniformemente distriuido entre 0 y 1 , luego hallamos el numero aleatorio

$$x=a+R(b-a)$$

Para $R=0.8594$ obtenemos $x=5.5782$

**Solución:**


```
R=0.8594
x=3+R*(6-3)
x
```

```
R=runif(1,0,1)
R
x=3+R*(6-3)
x
```

**Ejemplo:**

Para obtener un valor aleatorio de la distribución exponencial

$$
f(x)=\left\{
\begin{array}{ll}
\lambda e^{-\lambda x} , \ si \ x \geq 0 \\
0 \ , \  \ si \ x < 0
\end{array}
\right.
$$


Aplicando la transformada inversa

1. Hallamos $F(x)$



$$
F(x)=1-e^{-\lambda x}
$$

2. Generamos el número aleatorio  $R$


$$R \thicksim U(0,1)$$


3. $R=F(x)$
    
$$
R=1-e^{-\lambda x}\\
x=-\frac{1}{\lambda}ln(1-R)
$$

```
set.seed(111)
R=runif(1000)
x=-(log(1-R))/3
hist(x, freq=FALSE)
curve(dexp(x,3),add=TRUE)
```

```
set.seed(121)
x=rexp(1000,3)
hist(x, freq=FALSE,col="green3")
curve(dexp(x,3),add=TRUE,lwd=3,col="blue")
```

## **Método de convolucion**

Este método consiste en expresar la muestra deseada como la suma estadística de otras varaiables aleatorias

1. Una variable con distribución normal puede obtenerse sumando un numero suficiente de variables aleatorias con cualquier distribución.
Segun el teorema del limite central la suma(convolucuón) de $n$ variables aleatorias independientes identicamente distribuidas se hace asintóticamnete normal a medida que $n \longrightarrow \infty $

$$X=R_1+R_2+...+R_n$$

$$y=\mu +\sigma z$$

  donde $$z=\frac{x-\frac{n}{2}}{\sqrt{\frac{n}{12}}}$$

```
n=24
#set.seed(214)
set.seed(244)
R=runif(n,0,1)
R=round(R,4)
R
```

```
sum(R)
```

```
n=24
set.seed(244)
R=runif(n,0,1)
Mx=15  #25
Dx=1.3  #3
x=sum(R)
z=(x-n/2)/(sqrt(n/12))
y=Mx+Dx*z
y
```

2. Una variable aleatoria de Erlang se define como la suma de $m$ variables aleatorias exponenciales independientes identicamente distribuidas

$$Y=Y_1+Y_2+...+Y_n$$

donde $Y_i$ es exponencial

$$f(y_i)=\lambda e^{-\lambda y_i}, y_i\geq 0,  i=1,...,m$$

$$Y=-\frac{1}{\lambda}Ln\prod_{i=1}^n R_i$$


```
n=5  #4
set.seed(233)  #213
R=runif(n,0,1)
R=round(R,4)
L=3  #5
P=prod(R)
y=(-1/L)*log(P)
y
```
