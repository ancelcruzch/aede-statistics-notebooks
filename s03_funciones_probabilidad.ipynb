Para cada distribución, R dispone de cuatro funciones. Se puede acceder a cada una de ellas simplemente precediendo el nombre de la distribución en cada una de las tablas anteriores por la letra que se indica a continuación:

- **d:** función de densidad o de probabilidad.

- **p:** función de distribución

- **q:** función para el cálculo de cuantiles.

- **r:** función para simular datos con dicha distribución

**Ejemplo:** Para la distribución Binomial

Si $X$ sigue una distribución binomial $B(n,p)$, entonces

- **dbinom** proporciona la función de densidad

- **pbinom** proporciona la función de distribución de probabilidad acumulada del lado izquierdo

- **qbinom** proporciona la función de cuantiles

- **rbinom** genera valores aleatorios.

$dbinom(k,n,p)= P[X=k]$

$pbinom(k,n,p)=P[X \leq k]$

$qbinom(a,n,p)= min \{ x:P[X \leq x] \geq a \} $

$rbinom(m,n,p)$


**Ejemplo:** Considere un experimento Binomial con $n=10$ y $p=0.4$, entonces la función de distribución de probabilidad es

\begin{equation}
f(x)= {10\choose x} 0.4^x0.6^{10-x}\ con\ x=0,1,2,3,...,10
\end{equation}

 y su grafica correspondiente se obtiene como sigue

```r
x=seq(0,10)
y=dbinom(x,10,0.4)
plot(x,y,col="red",pch=19)
grid()
```

**Ejemplo:** Considere un experimento Binomial con $n=15$ y $p=0.8$, entonces la función de distribución de probabilidad es

\begin{equation}
f(x)= {15\choose x} 0.8^x0.2^{15-x}\ con\ x=0,1,2,3,...,15
\end{equation}

 y su grafica correspondiente se obtiene como sigue

## **Gráfica**

```r
x=seq(0,15,1)
y=dbinom(x,15,0.8)
plot(x,y, col="red")
```

**Calculo de probabilidades**

a) $P[X=12]$



```r
dbinom(12,15,0.8)
```

b) $P[X<= 12]$

```r
pbinom(12,15,0.8)
```

c) $P[X>7]=1-P[X<=7]$

```r
1-pbinom(7,15,0.8)
```

d) $P[4\leq X \leq 13]=P[X \leq 13]-P[X \leq 3]$

```r
pbinom(13,15,0.8)-pbinom(3,15,0.8)
```

**Ejemplo:** Para la distribución normal estandar (media 0 y desviación típica 1)


- **dnorm** proporciona la función de densidad

- **pnorm** proporciona la función de distribución de probabilidad acumulada del lado izquierdo

- **qnorm** proporciona la función de cuantiles

- **rnorm** genera valores aleatorios.


$dnorm(x, mean = 0, sd = 1)$

$pnorm(q, mean = 0, sd = 1, lower.tail = TRUE)$

$qnorm(p, mean = 0, sd = 1, lower.tail = TRUE)$

$rnorm(n, mean = 0, sd = 1)$


**Ejemplo:** Considere la distribución normal con media $\mu=70$ y $\sigma =4$, entonces siendo su función de densidad

\begin{equation}
f(x)= \frac{1}{\sqrt{2 \pi}\sigma}e^{-\frac{1}{2}\left(\frac{x-\mu}{\sigma} \right) ^{2}}  \ con\ x \in \mathbb{R}         
\end{equation}

su grafica se obtiene como sigue

```r
x=seq(54,86,0.1)
y=dnorm(x,mean=70,sd=4)
plot(x,y,type='l',col="blue",lwd=2)
grid()
```
