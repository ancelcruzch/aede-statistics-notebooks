# **Variable aleatoria continua**

Es una variable cuyo rango es un conjunto no numerable


# **Esperanza Matematica**


$$\mu=E(X)=\displaystyle\int_{-\infty}^{\infty}xf (x)dx$$

# **Varianza**


$$\sigma^2=V(X)=\displaystyle\int_{-\infty}^{\infty}(x-\mu)^2f (x)dx$$

# **Distribución de probabilidad para una variable aleatoria continua**

Si $X$ es una variable aleatoria continua, su función de densidad, debe verificar las siguientes condiciones

a)	$f (x)\geq  0  \        \forall x \in \mathbb{R}$

b)	$ \displaystyle\int_{-\infty}^{\infty}f (x)dx =1 $   , con   $x \in \mathbb{R}$

Si $E$ es el evento dado por $E=[a;b]$

$$P(E)=P[a \leq X \leq b]=\displaystyle\int_a^bf (x)dx$$

# **Distribucion uniforme**

$$
f(x)=\left\{
\begin{array}{ll}
\frac{1}{b-a} , \ si \ a \leq x \leq b\\
0 \ , \  en \ otro \ caso
\end{array}
\right.
$$

Su función de distribución acumulada esta dada por


$$
F(x)=P[X \leq x]=\int_{-\infty}^x f(t)dt=\left\{
\begin{array}{lll}
0, \ si \ x < a\\
\frac{x-a}{b-a} , \ si \ a \leq x < b\\
1 \ , \  si \ x \geq b
\end{array}
\right.
$$


$\mu= E[X]=\frac{a+b}{2}$

$\sigma^2=V[X]=\frac{{(b-a)}^2}{12}$

Para representar que una variable aleatoria $X$ sigue una distribución uniforme de parámetros $a$ y $b$, se denota:

\begin{equation}
X  \thicksim U(a,b)
\end{equation}

# **Distribucion exponencial**

Una variable aleatoria $X$ tiene una distribución exponencial si su función de densidad es


$$
f(x)=\left\{
\begin{array}{ll}
\lambda e^{-\lambda x} , \ si \ x \geq 0 \\
0 \ , \  \ si \ x < 0
\end{array}
\right.
$$

donde

$\mu =E[X]=\frac{1}{\lambda}$

$\sigma ^2 =V[X]= \frac {1}{\lambda^2}$


$$
F(x)=P[X \leq x]=\left\{
\begin{array}{ll}
1- e^{-\lambda x} , \ si \ x \geq 0 \\
0 \ , \  \ si \ x < 0
\end{array}
\right.
$$

Para representar que una variable aleatoria $X$ sigue una distribución exponencial de parámetro $\lambda$ , se denota:

\begin{equation}
X  \thicksim Exp(\lambda)
\end{equation}

 $$

$$

$V[X]=\frac{16}{12}$
