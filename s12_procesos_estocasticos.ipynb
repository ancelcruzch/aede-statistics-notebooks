# **Proceso estocastico**

Sea un espacio muestral $\Omega$, Si para cada resultado posible, $\omega$, tenemos una observación del fenómeno dada por una función real, $x(t,\omega)$ con $t\in I \subset \mathbb R $, para cada $\omega$ denominamos a $x(t,\omega)$ realización o función muestral

Se denomina proceso estocastico al conjunto

$$\left\{ X(t,\omega): t \in I, \omega \in \Omega \right\}$$

observese que para cada $t_0 \in I$, $X(t_0,\omega)$ es una variable aleatoria


Podemos decir que un proceso estocastico es un afuncion definida por

$$X: \Omega \rightarrow \mathbb F$$
$$\omega \mapsto x(t,\omega)$$

donde $\mathbb F$ es una familia de funciones definida en $t \in I$

Para un valor fijo $\omega_i$ , el proceso estocastico $x(t,\omega_i)$ representa una función

**Variable aleatoria**

Si $t_0 \in I \subset \mathbb R $,  $X(t_0,\omega)$ es una variable aleatoria

$$X: \Omega \rightarrow \mathbb R$$



**Vector aleatorio**

Si se fijan $n$ valores $t_1,t_2,...,t_n$ se obtiene un vector aleatorio

$$
\begin{equation}
\begin{bmatrix}
x(t_1,\omega)\\
x(t_2,\omega)\\
\vdots\\
x(t_n,\omega)]
\end{bmatrix}
\end{equation}$$

$$ $$

$$X: \Omega \rightarrow \mathbb R^n$$

**Proceso aleatorio**

$$X: \Omega \rightarrow \left\{ funciones \ reales \ de \ variable \ real \right\}$$

**Ejemplo:** Sea $\omega$ una variable aleatoria uniforme definida en $<-1,1>$. Definimos el proceso en tiempo continuo $X(t,\omega)$ como

$$X(t,\omega)=\omega cos(2\pi t)$$

sus funciones muestrales son ondas sinusoidales de amplitud aleatoria en $<-1,1>$.

```
w=c(0.2,0.7,1)
w
```

```
w[1]
```

```
w=c(0.2,0.7,1)
t=seq(0,2,0.01)
xt1=w[1]*cos(2*pi*t)
xt2=w[2]*cos(2*pi*t)
xt3=w[3]*cos(2*pi*t)
plot(t,xt1,type='l',xlim=c(0,2.5),ylim=c(-1,1),col="blue")
lines(t,xt2,type='l',col="red")
lines(t,xt3,type='l',col="green")
grid()
legend("topright", c("X(t,w1); w1=0.2","X(t,w2); w2=0.7", "X(t,w3); w3=1"),
       lty = 1, col = c("blue", "red", "green"))
```

```
w=c(0.2,0.7,1)
t=seq(0,2,0.01)
xt1=w[1]*cos(2*pi*t)
xt2=w[2]*cos(2*pi*t)
xt3=w[3]*cos(2*pi*t)
par(mfrow=c(3,1))
plot(t,xt1,type='l',ylim=c(-1,1),col="blue")
grid()
legend("topright", c("X(t,w1); w1=0.2"))
plot(t,xt2,type='l',col="red")
grid()
legend("topright", c("X(t,w2); w2=0.7"))
plot(t,xt3,type='l',col="green")
grid()
legend("topright", c("X(t,w3); w3=1"))
```

**Ejemplo:** Sea $\omega$ una variable aleatoria uniforme definida en $<-\pi,\pi>$. Definimos el proceso en tiempo continuo $X(t,\omega)$ como

$$X(t,\omega)=cos(2\pi t+\omega)$$

sus funciones muestrales son traslaciones aleatorias de la función $cos(2\pi t)$.

```
w=c(-0.3,0.3,3.4)
t=seq(-2,2,0.01)
xt1=cos(2*pi*t+w[1])
xt2=cos(2*pi*t+w[2])
xt3=cos(2*pi*t+w[3])
plot(t,xt1,type='l',xlim=c(-2.5,2.5),ylim=c(-1,1),col="blue")
lines(t,xt2,type='l',col="red")
lines(t,xt3,type='l',col="green")
grid()
legend("topright", c("X(t,w1); w1=-0.3","X(t,w2); w2=0.3", "X(t,w3); w3=3.4"),
       lty = 1, col = c("blue", "red", "green"))
```
